```python
#Visualise the data points using 2 features
def scatter_plot(xl,yl):
    sns.scatterplot(data = df,x = xl,y = yl)
```

```python
def show_cluster(xl,yl,zl,cent):
    sns.scatterplot(data = df,x = xl,y = yl, hue = zl,palette="deep")
    sns.scatterplot(data = cent,x = 'xc', y = 'yc',s=100, color=".2", marker="+" )
```

```python
def initial_centroids(x,y,k):
    rl = random.sample(range(0, 199), k)
    centroids = []
    i = 0
    while i<k:
        l = []
        l.append(x[rl[i]])
        l.append(y[rl[i]])
        centroids.append(l)
        i+=1
    return centroids
```

```python
def assign_cluster(x,y,cluster,centroids):
    i = 0
    k = len(centroids)
    ch = 0
    while(i<200):
        j = 0
        cm = pow(10,9)
        while j<k:
            cd = np.sqrt(pow(centroids[j][0]-x[i],2)+pow(centroids[j][1]-y[i],2))
            if cd<cm:
                cm = cd
                cl = j+1
            j+=1
        if(cluster[i]!=cl):
            cluster[i] = cl
            ch+=1
        i+=1
    return ch
```

```python
def update_centroids(x,y,cluster,centroids):
    i = 0
    l = len(centroids)
    while i<l:
        cnt = 0
        cx = 0
        cy = 0
        j = 0
        while j<200:
            if(cluster[j]==i+1):
                cnt+=1;
                cx+=x[j]
                cy+=y[j];
            j+=1
        centroids[i][0] = cx/cnt
        centroids[i][1] = cy/cnt
        i+=1
        
```

```python
xl = 'Annual_Income'
yl = 'Spending_Score'
x = df[xl].to_list()
y = df[yl].to_list()
cluster = [0 for cc in range(200)]
```

```python
k = 5
centroids = initial_centroids(x,y,k)
```

```python
centroids
```

```python
ct = 0
while 1:   #infinite loop
    t = assign_cluster(x,y,cluster,centroids)
    if(t==0):
        break
    update_centroids(x,y,cluster,centroids)
    ct+=1
```

```python
print("Number of iterations is equal to",ct)
```

```python
df['Cluster'] = cluster
```
